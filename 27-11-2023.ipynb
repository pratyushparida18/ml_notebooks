### Blending Techniques

#### addWeighted()

The cv2.addWeighted() function in OpenCV is used for blending two images together based on a specified weight. It computes the weighted sum of two images, allowing you to control the contribution of each image to the final result. The formula for the operation is:

dst(x,y)=alpha⋅img1(x,y)+beta⋅img2(x,y)+gamma

Here, 
img1
img1 and 
img2
img2 are the input images, and 
alpha
alpha, 
beta
beta, and 
gamma
gamma are the weights and the scalar added to the sum.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load two images
img1 = cv2.imread('/home/pratyush/Downloads/TestImg.jpg')
img2 = cv2.imread('/home/pratyush/Downloads/download.jpeg')

# Make sure the images have the same size
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Specify the weights and gamma value
alpha = 0.7
beta = 0.3
gamma = 0

# Perform the weighted sum
blended = cv2.addWeighted(img1, alpha, img2, beta, gamma)

# Display the images using Matplotlib
plt.figure(figsize=(10, 8))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Image 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))
plt.title('Blended')
plt.axis('off')

plt.show()
```

#### cv2.add()

This function adds pixel values of two images. It performs per-element addition without scaling.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load two images
img1 = cv2.imread('/home/pratyush/Downloads/TestImg.jpg')
img2 = cv2.imread('/home/pratyush/Downloads/download.jpeg')

# Make sure the images have the same size
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Add the images
added = cv2.add(img1, img2)

# Display the images using Matplotlib
plt.figure(figsize=(10, 8))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Image 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(added, cv2.COLOR_BGR2RGB))
plt.title('Added')
plt.axis('off')

plt.show()
```

#### cv2.divide()

#### cv2.bitwise_and(img1, img2):
Performs a bitwise AND operation on the corresponding pixel values of two images.
The result is 1 where both pixel values are 1; otherwise, it's 0.

#### cv2.bitwise_or(img1, img2):
Performs a bitwise OR operation on the corresponding pixel values of two images.
The result is 1 if at least one of the pixel values is 1.

#### cv2.bitwise_xor(img1, img2):
Performs a bitwise XOR (exclusive OR) operation on the corresponding pixel values of two images.
The result is 1 if the pixel values are different; otherwise, it's 0.

#### cv2.bitwise_not(img):
Performs a bitwise NOT operation on the pixel values of a single image.
The result is the bitwise complement of the input image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create two simple images
img1 = np.zeros((100, 100), dtype=np.uint8)
img2 = np.zeros((100, 100), dtype=np.uint8)

# Add a white rectangle to the first image
cv2.rectangle(img1, (10, 10), (90, 90), 255, -1)

# Add a white circle to the second image
cv2.circle(img2, (50, 50), 40, 255, -1)

# Bitwise AND operation
and_result = cv2.bitwise_and(img1, img2)

# Bitwise OR operation
or_result = cv2.bitwise_or(img1, img2)

# Bitwise XOR operation
xor_result = cv2.bitwise_xor(img1, img2)

# Bitwise NOT operation (on img1 for demonstration)
not_result = cv2.bitwise_not(img1)

# Display the results
plt.figure(figsize=(8, 8))

plt.subplot(2, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(img2, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(and_result, cmap='gray')
plt.title('Bitwise AND')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(or_result, cmap='gray')
plt.title('Bitwise OR')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(xor_result, cmap='gray')
plt.title('Bitwise XOR')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(not_result, cmap='gray')
plt.title('Bitwise NOT (on Image 1)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

### Gaussian Pytramid and Laplacian Pyramid

#### Gaussian Pyramid:
    
Construction:
Downsampling: The Gaussian pyramid is constructed by applying a series of low-pass Gaussian filters to the original image and then subsampling it. This process is repeated to create multiple levels or layers of the pyramid.
Each level of the pyramid represents a progressively downsampled and blurred version of the original image.

Purpose:
Image Smoothing and Downsampling: The primary purpose of the Gaussian pyramid is to create a multi-scale representation of an image. Each level contains information at a different scale, with lower levels capturing more global features and higher levels containing finer details.

The operation to generate each level i of the Gaussian pyramid (Gi) is often denoted as pyrDown(Gi-1), where Gi-1 is the previous level. Gi = pyrDown(Gi-1)
The downsampling is typically achieved by convolving the image with a Gaussian filter and then subsampling.

#### Laplacian Pyramid

Upsampling and Subtraction: 
The Laplacian pyramid is constructed based on the levels of the Gaussian pyramid. Each level of the Laplacian pyramid (Li) is obtained by subtracting an upsampled and blurred version of the image from the corresponding level of the Gaussian pyramid.

Li=Gi−pyrUp(Gi)
The subtraction of the upsampled image helps retain the details that were removed during the downsampling process.

Purpose:
Image Reconstruction and Detail Enhancement: The Laplacian pyramid represents the details that were lost during the construction of the Gaussian pyramid. It is used for reconstructing the original image and enhancing fine details.

Mathematically:
The Laplacian pyramid is constructed using the difference between each level of the Gaussian pyramid and an upsampled and blurred version of the next level.

Li = Gi - pyrUp(Gi+1)

```python
import cv2
import numpy as np
apple = cv2.imread('apple.jpg')
orange = cv2.imread('orange.jpg')
print(apple.shape)
print(orange.shape)
apple_orange = np.hstack((apple[:, :256], orange[:, 256:]))

# generate Gaussian pyramid for apple
apple_copy = apple.copy()
gp_apple = [apple_copy]
for i in range(6):
    apple_copy = cv2.pyrDown(apple_copy)
    gp_apple.append(apple_copy)


# generate Gaussian pyramid for orange
orange_copy = orange.copy()
gp_orange = [orange_copy]
for i in range(6):
    orange_copy = cv2.pyrDown(orange_copy)
    gp_orange.append(orange_copy)

# generate Laplacian Pyramid for apple
apple_copy = gp_apple[5]
lp_apple = [apple_copy]
for i in range(5, 0, -1):
    gaussian_expanded = cv2.pyrUp(gp_apple[i])
    laplacian = cv2.subtract(gp_apple[i-1], gaussian_expanded)
    lp_apple.append(laplacian)

# generate Laplacian Pyramid for orange
orange_copy = gp_orange[5]
lp_orange = [orange_copy]
for i in range(5, 0, -1):
    gaussian_expanded = cv2.pyrUp(gp_orange[i])
    laplacian = cv2.subtract(gp_orange[i-1], gaussian_expanded)
    lp_orange.append(laplacian)

# Now add left and right halves of images in each level
apple_orange_pyramid = []
n = 0
for apple_lap, orange_lap in zip(lp_apple, lp_orange):
    n += 1
    cols, rows, ch = apple_lap.shape
    laplacian = np.hstack((apple_lap[:, 0:int(cols/2)], orange_lap[:, int(cols/2):]))
    apple_orange_pyramid.append(laplacian)
# now reconstruct
apple_orange_reconstruct = apple_orange_pyramid[0]
for i in range(1, 6):
    apple_orange_reconstruct = cv2.pyrUp(apple_orange_reconstruct)
    apple_orange_reconstruct = cv2.add(apple_orange_pyramid[i], apple_orange_reconstruct)

cv2.imshow("apple", apple)
cv2.imshow("orange", orange)
cv2.imshow("apple_orange", apple_orange)
cv2.imshow("apple_orange_reconstruct", apple_orange_reconstruct)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Usecases

cv2.add():
Use Case: Addition of pixel values of two images element-wise.
Application: Image brightness adjustment, combining images with different intensity levels.

cv2.subtract():
Use Case: Subtraction of pixel values of one image from another element-wise.
Application: Image differencing, highlighting changes between images.

cv2.multiply():
Use Case: Per-element multiplication of pixel values of two images.
Application: Contrast adjustment, image masking.

Use Case: Per-element division of pixel values of one image by another.
Application: Image normalization, color balance correction.

cv2.addWeighted():
Use Case: Blending two images using a weighted sum.
Application: Image composition, creating visual effects, smooth transitions.

##### What is the effect on resolution by these pyramids

Gaussian Pyramid:
Downsampling:

Each level of the Gaussian Pyramid is obtained by smoothing and then subsampling the image from the previous level.
Downsampling reduces the spatial resolution of the image, making it smaller in dimensions.
Effect on Image Resolution:

As you move up the Gaussian Pyramid, the images become progressively smaller, losing finer details.
The higher levels of the pyramid represent lower resolution and more globally averaged versions of the original image.
Laplacian Pyramid:
Upsampling:

Each level of the Laplacian Pyramid is created by taking the difference between an upsampled and blurred version of the image from the next lower level and the image at that level.
Effect on Image Resolution:

The Laplacian Pyramid contains details that were lost during the construction of the Gaussian Pyramid.
Upsampling in the Laplacian Pyramid process helps retain and enhance fine details, providing a mechanism for reconstructing the original image with more details.
Overall Impact:
Resolution Reduction:

Gaussian Pyramid leads to a reduction in resolution as you move up the pyramid.
The original image's details are progressively lost, and the image becomes coarser.
Resolution Enhancement:

Laplacian Pyramid, in combination with Gaussian Pyramid, allows for reconstructing the original image with enhanced details.
The upsampling process in the Laplacian Pyramid helps recover and refine fine details.
Multi-Resolution Representation:

The combination of Gaussian and Laplacian Pyramids provides a multi-resolution representation of the image, where each level captures different scales of features.
