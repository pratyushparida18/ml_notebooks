### Sobel Operator

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
image = cv2.imread('/home/pratyush/Downloads/edge.jpg', cv2.IMREAD_GRAYSCALE)

blurred = cv2.GaussianBlur(image, (3, 3), 0)

# Apply Sobel operator
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Compute gradient magnitude
gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Normalize the gradient magnitude to the range [0, 255]
gradient_magnitude_normalized = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Convert to 8-bit unsigned integer
gradient_magnitude_uint8 = np.uint8(gradient_magnitude_normalized)

# Plot the original image and Sobel results using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(gradient_magnitude_uint8, cmap='gray')
plt.title('Sobel Operator')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.show()
```

#### Working


Convolution Operation:
The Sobel operator uses convolution with a small kernel (a matrix) to compute the gradient of the image intensity.
It involves sliding a small kernel over the image and computing the sum of element-wise products between the kernel and the corresponding region in the image.

Gradient Computation:
The Sobel operator uses two 3x3 kernels, one for horizontal changes and the other for vertical changes. These kernels are designed to detect changes in intensity in the horizontal and vertical directions.
The horizontal Sobel kernel typically looks like this:
-1  0  1
-2  0  2
-1  0  1

The vertical Sobel kernel is the transpose of the horizontal Sobel kernel.

Gradient Magnitude and Direction:
After applying the Sobel operator in both horizontal and vertical directions, the gradient magnitude and direction can be computed.
The magnitude is calculated as the square root of the sum of squares of the horizontal and vertical gradients.
The direction is the arctangent of the ratio of the vertical gradient to the horizontal gradient.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
image = cv2.imread('/home/pratyush/Downloads/TestImg.jpg', cv2.IMREAD_GRAYSCALE)

blurred = cv2.GaussianBlur(image, (3, 3), 0)

# Apply Sobel operator
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Compute gradient magnitude
gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Normalize the gradient magnitude to the range [0, 255]
gradient_magnitude_normalized = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Convert to 8-bit unsigned integer
gradient_magnitude_uint8 = np.uint8(gradient_magnitude_normalized)

# Plot the original image and Sobel results using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(gradient_magnitude_uint8, cmap='gray')
plt.title('Sobel Operator')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.show()
```

#### Scharr Operator

The Scharr operator is an edge detection operator that is often used as an alternative to the Sobel operator. It is designed to provide better rotational symmetry and is particularly useful for detecting edges in images. The Scharr operator is implemented as a convolution operation with a 3x3 kernel.

The Scharr operator kernels for detecting image gradients in the x and y directions are given by:
Scharr_x = 
[
 -3, 0, 3;
-10, 0, 10;
 -3, 0, 3
]

Scharr_y = 
[
 -3, -10, -3;
  0,   0,  0;
  3,  10,  3
]


#### Working
Image Gradients in X and Y Directions:
The Scharr operator is applied to the image in both the x and y directions using convolution with the specified kernels.
The convolution involves sliding the Scharr kernels over the image and computing the sum of element-wise products at each position.

Gradient Magnitude:
The gradient magnitude at each pixel is computed as the square root of the sum of squared gradients in both the x and y directions.
Magnitude = sqrt((Scharr_x)^2 + (Scharr_y)^2)
 
Normalization:
The gradient magnitude is often normalized to a specific range for better visualization.

#### Laplacian Operator

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
image = cv2.imread('/home/pratyush/Downloads/edge.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur for noise reduction
blurred = cv2.GaussianBlur(image, (7, 7), 0)

# Apply Laplacian operator
laplacian = cv2.Laplacian(blurred, ksize=5, ddepth=cv2.CV_16S)

#converting back into 8 bit format for proper visualization with matplotlib
filtered_image = cv2.convertScaleAbs(laplacian)


# Plot the original image, blurred image, and Laplacian result using matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title('Blurred Image (Gaussian Filter)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(filtered_image, cmap='gray')
plt.title('Laplacian Operator after Gaussian Blur')
plt.axis('off')

plt.show()
```

The Laplacian operator is a 2D spatial filter that is commonly used for edge detection in image processing. It calculates the second spatial derivative of an image, emphasizing regions of rapid intensity change. The Laplacian operator is often applied after smoothing an image to reduce noise.

In discrete form, the Laplacian operator is usually implemented as a convolution operation with a convolution kernel. Commonly used 3x3 kernels for the Laplacian operator include:

[ 0  1  0 ]
[ 1 -4  1 ]
[ 0  1  0 ]

or

[ 1  1  1 ]
[ 1 -8  1 ]
[ 1  1  1 ]

Steps:

smoothing :
Before applying the Laplacian operator, the image is often smoothed using a Gaussian filter to reduce noise. This helps to prevent the detection of noise as edges.

Gradient Calculation:
The Laplacian operator is applied to the image using convolution. The convolution process involves sliding the Laplacian kernel over each pixel of the image and computing the sum of products.

Edge Enhancement:
The Laplacian operator highlights areas where the intensity changes rapidly, emphasizing edges and fine details in the image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
image = cv2.imread('/home/pratyush/Downloads/TestImg.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur for noise reduction
blurred = cv2.GaussianBlur(image, (7, 7), 0)

# Apply Laplacian operator
laplacian = cv2.Laplacian(blurred, ksize=5, ddepth=cv2.CV_16S)

#converting back into 8 bit format for proper visualization with matplotlib
filtered_image = cv2.convertScaleAbs(laplacian)

# Plot the original image, blurred image, and Laplacian result using matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title('Blurred Image (Gaussian Filter)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(filtered_image, cmap='gray')
plt.title('Laplacian Operator after Gaussian Blur')
plt.axis('off')

plt.show()
```

#### Canny Edge Detection

```python
import cv2
import matplotlib.pyplot as plt

# Read the input image
image = cv2.imread('/home/pratyush/Downloads/edge.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur for noise reduction
blurred = cv2.GaussianBlur(image, (3, 3), 0)

# Apply Canny edge detection
edges = cv2.Canny(blurred, 50, 150)

# Plot the original image, blurred image, and Canny edges using matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title('Blurred Image (Gaussian Filter)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.show()
```

#### Working

Canny edge detection is a multi-stage algorithm that aims to identify edges in an image. Here are the main steps of the Canny edge detection algorithm:

Smoothing (Gaussian Blur):
The first step is to apply a Gaussian blur to the image to reduce noise. This is crucial because the Canny algorithm is sensitive to noise, and blurring helps to smooth out the image.

Gradient Calculation:
Calculate the gradient of the image intensity. This involves finding the intensity gradients in both the x and y directions. The gradient magnitude and direction are computed using the following formulas:

Magnitude = sqrt(G_x^2 + G_y^2)

Direction = atan2(G_y, G_x)


Non-Maximum Suppression:
Suppress non-maximum gradients by keeping only the local maxima in the gradient direction. This helps thinning the edges to one-pixel wide lines.

Edge Tracking by Hysteresis:
Threshold the gradient magnitude to identify potential edge pixels. There are two thresholds: a low threshold (low_threshold) and a high threshold (high_threshold).
Pixels with gradient magnitudes above the high threshold are considered strong edge pixels and are likely part of the final edge.
Pixels with gradient magnitudes below the low threshold are considered non-edge pixels and are discarded.
Pixels with gradient magnitudes between the low and high thresholds are considered weak edge pixels.
Weak edge pixels are included in the final edge output only if they are connected to strong edge pixels. This helps suppress noise and ensures that only significant edges are retained.

Edge Tracking and Final Thresholding:
Edge pixels are finalized based on connectivity. Weak edge pixels that are connected to strong edge pixels are included in the final edge output, while isolated weak edge pixels are discarded.

```python
import cv2
import matplotlib.pyplot as plt

# Read the input image
image = cv2.imread('/home/pratyush/Downloads/TestImg.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur for noise reduction
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Apply Canny edge detection
edges = cv2.Canny(blurred, 30, 40,L2gradient=None,apertureSize=3)

# Plot the original image, blurred image, and Canny edges using matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title('Blurred Image (Gaussian Filter)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the input image
img = cv2.imread('/home/pratyush/Downloads/TestImg.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

thresh = cv2.threshold(gray, 85, 255, cv2.THRESH_BINARY)[1]

kernel = np.ones((3, 3), np.uint8)
morph_result = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

edges = 255 - morph_result

# Perform Canny edge detection
canny_edges = cv2.Canny(edges, 30, 40)

# Apply morphological dilation
dilated_edges = cv2.dilate(canny_edges, kernel, iterations=1)

# Plot the original image, thresholded image, morphological result, and dilated Canny edges
plt.figure(figsize=(16, 4))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# plt.subplot(1, 4, 2)
# plt.imshow(thresh, cmap='gray')
# plt.title('Thresholded Image')
# plt.axis('off')

# plt.subplot(1, 4, 3)
# plt.imshow(edges, cmap='gray')
# plt.title('Morphological Operations')
# plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(dilated_edges, cmap='gray')
plt.title('Dilated Canny Edges')
plt.axis('off')

plt.show()
```
