```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Callback function for trackbar changes (does nothing in this example)
def nothing(x):
    pass

# Read the images
image1 = cv2.imread("/home/pratyush/Downloads/sample_shirt.jpg")
image2 = cv2.imread("/home/pratyush/Downloads/shirt2.jpg")

# Convert images from BGR to RGB
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# Create a window with sliders
cv2.namedWindow('Adjust Colors')

# Initialize sliders for lower and upper RGB values
cv2.createTrackbar('Lower_R', 'Adjust Colors', 0, 255, nothing)
cv2.createTrackbar('Lower_G', 'Adjust Colors', 0, 255, nothing)
cv2.createTrackbar('Lower_B', 'Adjust Colors', 0, 255, nothing)
cv2.createTrackbar('Upper_R', 'Adjust Colors', 255, 255, nothing)
cv2.createTrackbar('Upper_G', 'Adjust Colors', 255, 255, nothing)
cv2.createTrackbar('Upper_B', 'Adjust Colors', 255, 255, nothing)

while True:
    # Get current trackbar positions
    lower_r = cv2.getTrackbarPos('Lower_R', 'Adjust Colors')
    lower_g = cv2.getTrackbarPos('Lower_G', 'Adjust Colors')
    lower_b = cv2.getTrackbarPos('Lower_B', 'Adjust Colors')
    upper_r = cv2.getTrackbarPos('Upper_R', 'Adjust Colors')
    upper_g = cv2.getTrackbarPos('Upper_G', 'Adjust Colors')
    upper_b = cv2.getTrackbarPos('Upper_B', 'Adjust Colors')

    # Define the range of color in RGB
    lower_color = np.array([lower_r, lower_g, lower_b])
    upper_color = np.array([upper_r, upper_g, upper_b])

    # Create masks for the color range in both images
    mask1 = cv2.inRange(image1_rgb, lower_color, upper_color)
    mask2 = cv2.inRange(image2_rgb, lower_color, upper_color)

    # Apply the masks to the original images
    segmented_image1 = cv2.bitwise_and(image1_rgb, image1_rgb, mask=mask1)
    segmented_image2 = cv2.bitwise_and(image2_rgb, image2_rgb, mask=mask2)

    # Display the segmented images
    cv2.imshow('Segmented Image 1', cv2.cvtColor(segmented_image1, cv2.COLOR_RGB2BGR))
    cv2.imshow('Segmented Image 2', cv2.cvtColor(segmented_image2, cv2.COLOR_RGB2BGR))

    # Exit the loop when 'Esc' key is pressed
    if cv2.waitKey(1) & 0xFF == 27:
        break

# Destroy all OpenCV windows
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def sharpen_image(image):
    # Define a sharpening kernel
    kernel = np.array([[-1, -1, -1],
                       [-1,  9, -1],
                       [-1, -1, -1]])

    # Apply the sharpening kernel using filter2D
    sharpened_image = cv2.filter2D(image, -1, kernel)

    return sharpened_image

def color_correction(image):
    # Perform color correction using cv2.normalize
    normalized_image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)
    return normalized_image

def adjust_saturation(image, factor=1.5):
    # Convert the image to the HSV color space
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Adjust the saturation
    hsv_image[:, :, 1] = np.clip(hsv_image[:, :, 1] * factor, 0, 255).astype(np.uint8)

    # Convert the image back to BGR
    saturated_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)

    return saturated_image

def adjust_contrast(image, alpha, beta=0):
    # Adjust the contrast using cv2.convertScaleAbs
    contrast_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return contrast_image

mask2 = cv2.inRange(image2_hsv, lower_green, upper_green)

# Apply the mask to the original image
segmented_image2 = cv2.bitwise_and(image2, image2, mask=mask2)

# Apply the sharpening filter
sharpened_image = sharpen_image(segmented_image2)

# Perform color correction
corrected_image = color_correction(sharpened_image)

# Adjust saturation
final_image = adjust_saturation(corrected_image, factor=0.5)  

# Adjust contrast
final_image = adjust_contrast(final_image, alpha=1, beta=0)  # You can experiment with different alpha and beta values

# Blend the sharpened image on top of image2 using addWeighted
result_image = cv2.addWeighted(image2, 1, sharpened_image, 0.1, 0)

# Plot the original, sharpened, and final images using matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(segmented_image2, cv2.COLOR_BGR2RGB))
plt.title('Segmented Image 2')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.title('Result (Sharpened Image on top of Image 2)')
plt.axis('off')

cv2.imwrite("sharpened.jpg", result_image)

plt.show()
```
