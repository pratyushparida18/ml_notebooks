```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('/home/pratyush/Downloads/edge.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian Blur
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply Morphological Opening
kernel = np.ones((15, 15), np.uint8)
morph_open = cv2.morphologyEx(blurred, cv2.MORPH_OPEN, kernel)

# Apply Gaussian Blur again
final_result = cv2.GaussianBlur(morph_open, (1, 1), 0)

# Apply Canny edge detection with L2gradient=true
edges = cv2.Canny(final_result, 5, 200, apertureSize=3, L2gradient=False)

# Perform Morphological Closing
closing_kernel = np.ones((3, 3), np.uint8)
closed_edges = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, closing_kernel)

# Display the original image, the final result, the edges, and the closed edges
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(final_result, cmap='gray')
plt.title('After Operations')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection (L2gradient=True)')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(closed_edges, cmap='gray')
plt.title('Closed Edges')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/home/pratyush/Downloads/edge.jpg')

# Convert BGR to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the HSV range for yellow
lower_yellow = np.array([22, 0, 0])
upper_yellow = np.array([184, 255, 180])

# Create a mask
mask = cv2.inRange(hsv, lower_yellow, upper_yellow)
# Create a blank image of the same size as the original image
blank_image = np.zeros_like(image)

# Set the mask region to the desired value in the blank image
blank_image[np.where(mask != 0)] = (255, 255, 255)  
# Create the complement of the mask
inverse_mask = cv2.bitwise_not(blank_image)

# Bitwise AND between the original image and the complement of the mask
blurred_result = cv2.GaussianBlur(inverse_mask, (17, 17), 0)
# Convert the blurred result to grayscale
gray_blurred = cv2.cvtColor(blurred_result, cv2.COLOR_BGR2GRAY)

# Apply adaptive thresholding
adaptive_thresh = cv2.adaptiveThreshold(gray_blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 11)

# Apply Canny edge detection to the thresholded image
edges = cv2.Canny(adaptive_thresh, 50, 200)

# Find contours in the edges
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Sort contours by area (largest to smallest)
contours = sorted(contours, key=cv2.contourArea, reverse=True)

# Create a blank image to draw contours
contour_image = np.zeros_like(image)

# Draw the largest 50 contours
for i in range(min(50, len(contours))):
    cv2.drawContours(contour_image, [contours[i]], -1, (0, 255, 0), 2)

# Display the original image, edges, and the image with contours
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title('Contours (Top 50)')
plt.axis('off')

plt.show()
```
